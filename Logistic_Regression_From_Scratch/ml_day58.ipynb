```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
def step(z):
    return 1 if z>0 else 0
```

#### building logistc regression model using perceptron trick

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)  ##adding one extra column x0 (x.shape=(rows,3)
    weights = np.ones(X.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
        
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
# y=mx+b and ax+by+c=0
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
def perceptron(X,y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = perceptron(X,y)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(X,y):
        
    X = np.insert(X,0,1,axis=1)  ##adding one extra column x0 (x.shape=(rows,3)
    weights = np.ones(X.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept,coef=perceptron(X,y)
```

```python
m=-(coef[0]/coef[1])
b=-(intercept/coef[1])
```

```python
x_input2=np.linspace(-3,3,100)
y_input2=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label="perceptron_trick")
plt.plot(x_input1,y_input1,color='black',linewidth=3,label="sklearn_logistic_regression")
plt.plot(x_input2,y_input2,color='brown',linewidth=3,label="sigmoid")
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.legend()
plt.show()
```

-> sigmoid is good in comapre to step fuction but logistic still better
