```python
class MeraLR:
    def __init__(self):
        self.m=None
        self.b=None

    def fit(self,X_train,y_train): # to train model 
        num=0
        den=0
        for i in range(X_train.shape[0]): 
            num=num+((X_train[i]-X_train.mean())*(y_train[i]-y_train.mean()))      # x_train and y_train in array or numpy array form
            den=den+((X_train[i]-X_train.mean())*(X_train[i]-X_train.mean()))

        self.m=num/den
        self.b=y_train.mean()-(self.m*X_train.mean())
        print('m:',self.m)
        print('b:',self.b)

    def predict(self,X_test):
        return (self.m*X_test)+self.b

    def get_coef(self):
        return self.m

    def get_intercept(self):
        return self.b
```
